### What is Machine Learning?

As programmers, we're used to writing instructions based on different conditions - if this, then that. But for some problems it's really difficult to know what the rules are upfront. For example, let's say we want to predict the sale price of a house, given some data about the house, like number of bedrooms, number of bathrooms, size in sq. feet, and house age. We can write a function that takes each of these data points (known as **features**) as a parameter, puts a price on each quantity of a specific feature (like 10,000 dollars per bedroom), and returns the sum of the prices of all combined features. But how would we know how much to price each feature quantity? What we can do is, take a couple of houses that we know the sale price for and use them as examples. Try running your prediction function using several different combinations of values for the features, and see which combination makes the best predictions when compared to the actual sale price of the houses. So after your experimenting you may find that if you add 10k for each bedroom, 5k for each bathroom, 1k per sq foot, and subtract 3k for each year of the house's age, you get a fairly accurate prediction. You'd also probably want to add a base price to account for the minimum a house is worth. But manually figuring out what these values should be can be pretty tedious, even if we're only accounting for a few different features. Many problems require accounting for many features. So we need an automated way to figure out the best value to add for each feature.

In machine learning terms, the prices we decide to put on each feature are called **weights.** And every machine learning **model** from a simple house price predictor to ChatGPT, consists of weights like these, and **training** a model is the process by which we figure out what the value of these weights should be. This is the **learning** part of machine learning. We learn what the best weights are for the predictions we want to make.

There are a couple of aspects that make building a good ML model challenging, depending on the application. I'll list some of them here:

1. Which features should we use to make our predictions?
2. Some features are not in a format that would be useful for a weighted sum calculation (for example, zip codes) and we need to convert them into values that are useful.
3. For many applications, we don’t have useful features up-front, and we need to create useful features from whatever we do have. For applications like image recognition, and natural language processing, it’s not obvious what useful features are and there are several approaches for creating them. Some of these can get quite complicated.
4. We need to make sure that the data we use to train our model is statistically representative of the data we will be making the predictions on.
5. We need to know how to accurately determine whether our model works well even for data that we have never seen.

Don't worry about these for now. We'll cover them during the course of the series. They're just meant as a preview to show you that we have plenty of material to cover 😁. But the one thing to know for now is that the core of any ML model are its weights and what these weights should be applied to. If you understand this, you understand the basic premise of machine learning.

### **Exercises**
#### Implement the following functions, and ensure the tests pass.

```python
def predict(features, weights, base_price):
    """Predict the sale price of a house.

    Args:
        features: list of feature values, e.g. [sqft, bedrooms, bathrooms, age]
        weights: list of weights, same length as features
        base_price: base price before features are added
    Returns:
        predicted sale price
    """
    # YOUR CODE HERE
    return base_price + sum(x * y for x,y in zip(features, weights))
    raise NotImplementedError()


# --- Tests (do not modify) ---
# A 1500 sqft, 3 bed, 2 bath, 10 year old house
# with weights: $100/sqft, $10k/bed, $5k/bath, -$3k/year age
_price = predict([1500, 3, 2, 10], [100, 10000, 5000, -3000], 50000)
assert _price == 210000, f"Expected 210000, got {_price}"

# A 2000 sqft, 4 bed, 3 bath, 20 year old house
_price2 = predict([2000, 4, 3, 20], [100, 10000, 5000, -3000], 50000)
assert _price2 == 245000, f"Expected 245000, got {_price2}"

# Zero weights should return base price
_price3 = predict([1500, 3, 2, 10], [0, 0, 0, 0], 50000)
assert _price3 == 50000, f"Expected 50000, got {_price3}"

print("✅ All tests passed!")
```

```python
def prediction_error(predicted, actual):
    """How far off was our prediction?

    Args:
        predicted: predicted sale price
        actual: actual sale price
    Returns:
        the error (predicted - actual)
    """
    return predicted - actual
    raise NotImplementedError()


# --- Tests (do not modify) ---
assert prediction_error(200000, 300000) == -100000
assert prediction_error(350000, 300000) == 50000
assert prediction_error(300000, 300000) == 0
print("✅ All tests passed!")
```

Now let's make predictions on some data using several different weight sets, and determine which are the best weights to use.

```python
# Here are 4 houses we know the sale prices for:
houses = [
    [1500, 3, 2, 10],  # features: sqft, beds, baths, age
    [2000, 4, 3, 20],
    [800, 2, 1, 40],
    [2500, 5, 3, 5],
]
actual_prices = [300000, 400000, 150000, 500000]

# Try these three weight combinations (with base_price = 50000):
weight_sets = [
    [100, 10000, 5000, -3000],
    [150, 20000, 10000, -2000],
    [80, 15000, 8000, -1000],
]

def average_absolute_error(houses, actual_prices, weights, base_price):
    """Compute the average absolute error across all houses.

    Args:
        houses: list of feature lists
        actual_prices: list of actual sale prices
        weights: list of weights to try
        base_price: base price
    Returns:
        average absolute error (always positive)
    """
    # YOUR CODE HERE (use your predict and prediction_error functions)
    sum = 0
    for h, a in zip(houses, actual_prices):
      p = predict(h,weights,base_price)
      sum += abs(prediction_error(p, a))
    return sum/len(h)
    raise NotImplementedError()


# --- Tests (do not modify) ---
_err1 = average_absolute_error(houses, actual_prices, weight_sets[0], 50000)
_err2 = average_absolute_error(houses, actual_prices, weight_sets[1], 50000)
_err3 = average_absolute_error(houses, actual_prices, weight_sets[2], 50000)

assert isinstance(_err1, (int, float)), "Should return a number"
assert _err1 >= 0, "Absolute error should be non-negative"
assert _err2 >= 0, "Absolute error should be non-negative"
assert _err3 >= 0, "Absolute error should be non-negative"

# Errors that cancel out should NOT average to zero
_test_houses = [[1000, 2, 1, 5], [2000, 4, 2, 10]]
_test_prices = [200000, 230000]
_test_weights = [100, 10000, 5000, -3000]
_err = average_absolute_error(_test_houses, _test_prices, _test_weights, 50000)
print('_err', _err)
assert _err > 0, "Should compute absolute value of each error, not absolute value of the sum"

# Find the best weight set
_best = weight_sets[[_err1, _err2, _err3].index(min(_err1, _err2, _err3))]
assert _best == [150, 20000, 10000, -2000], f"Expected weight set 2 to be best, got {_best}"

print(f"Weight set 1 avg error: ${_err1:,.0f}")
print(f"Weight set 2 avg error: ${_err2:,.0f}")
print(f"Weight set 3 avg error: ${_err3:,.0f}")
print(f"Best weights: {_best}")
print("✅ All tests passed!")
```
